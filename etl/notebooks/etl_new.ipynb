```python
def load_source(a):
    fn = f'../source/{a}.zip'
    zf = zipfile.ZipFile(fn)
    res = dict()
    for f in zf.filelist:
        if 'WSHOCK' in f.filename:
            res['with_shock'] = pd.read_csv(BytesIO(zf.read(f.filename)))
        if 'NOSHOCK' in f.filename:
            res['no_shock'] = pd.read_csv(BytesIO(zf.read(f.filename)))
    if 'with_shock' not in res.keys() or 'no_shock' not in res.keys():
        fl = [f.filename for f in zf.filelist]
        raise ValueError(f'{a}: {fl}')
        
    return res
```

```python
def create_datapoint(df, indicator_name, measure_id, varient):
    res = df.loc[df['measure_id'] == measure_id,
                 ['location_id', 'sex_id', 'age_group_id', 'year_id',
                  'val', 'lower', 'upper']].copy()
    if res.empty:
        raise ValueError('empty dataframe')
    
    id_head = indicator_name + '_' + varient
    res.columns = ['location', 'sex', 'age_group', 'year', 
                   id_head, id_head+'_lower', id_head+'_upper']
    
    return res.set_index(['location', 'sex', 'age_group', 'year']).sort_index()
```

```python
create_datapoint(data['10']['with_shock'], 'lex', 26, 'with_shock')
```

```python
# only include lex:
# with shock: measure = 31
# no shock: measure = 26

datapoints = {'with_shock': [], 'no_shock': []}

for _, d in data.items():
    datapoints['with_shock'].append(create_datapoint(d['with_shock'], 'lex', 26, 'with_shock'))
    datapoints['no_shock'].append(create_datapoint(d['no_shock'], 'lex', 31, 'no_shock'))
```

```python
def create_entity(cols):
    res = []
    for _, v in data.items():
        res.append(v['with_shock'][cols].drop_duplicates())
        res.append(v['no_shock'][cols].drop_duplicates())
        
    return pd.concat(res, ignore_index=True).drop_duplicates()
```
